```r
# library
library(ggplot2)

tableData <- read.csv("InfluenzaPositiveTests.csv")

week <-c(tableData$Week)

colr <- c(rep("totalB",31),rep("totalA",31))

totalA <- tableData$TotalA
totalB <- tableData$TotalB

values <- c(totalB,totalA)
type <-c(totalB,totalA)
data <- data.frame(week,values)


plodata <- ggplot(data, aes(factor(week),values))+geom_bar(stat = "identity",aes(fill=factor(colr)))+
  scale_fill_manual(values=c("#FFE92F","#89E200"))

plodata


```
